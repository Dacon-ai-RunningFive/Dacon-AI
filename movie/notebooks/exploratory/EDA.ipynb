```python
import pandas as pd
import plotly.express as px
```

```python
m = pd.read_csv("../../data/processed/KOBIS_개봉일람_3.csv")

m
```

```python
# 상관분석
c = m[["전국 스크린수", "전국 매출액", "전국 관객수", "서울 매출액", "서울 관객수", "감독_로그", "제작사_로그", "배급사_로그"]]

c.corr()
```

```python
# 그래프 그리기
fig = px.scatter(m, x="전국 매출액", y="전국 스크린수",
                 color="전국 스크린수", color_continuous_scale=px.colors.sequential.Viridis)

fig.show()
```

```python
# 그래프 그리기
fig = px.bar(m, y="장르", x="전국 매출액", color="장르", color_discrete_sequence=px.colors.qualitative.G10)

fig.show()
```

```python
fig = px.pie(m, values="전국 매출액", names="등급")

fig.show()
```
