```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
m = pd.read_csv("src/KOBIS_개봉일람_ap.csv")

m
```

```python
감독 = m[m['감독'].isna()]['영화명'].tolist()
제작사 = m[m['제작사'].isna()]['영화명'].tolist()
배급사 = m[m['배급사'].isna()]['영화명'].tolist()
```

```python
url = "https://search.naver.com/search.naver?ie=UTF-8&query=" + "체이싱" + " 감독"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")
print(soup.select_one("#mflick > div > div > ul > li > div > div.title_box > strong").text)
```

```python
import requests
from bs4 import BeautifulSoup

director_names = {}

for title in tqdm(감독):
    url = "https://search.naver.com/search.naver?ie=UTF-8&query=" + title + " 감독"
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")
    try:
        director_name = soup.select_one("#mflick > div > div > ul > li > div > div.title_box > strong").text
        director_names[title] = director_name.strip()
    except:
        director_names[title] = "None"
    
director_names
```

```python
for k, v in director_names.items():
    m.loc[m["영화명"] == k, "감독"] = v
    print(m.loc[m["영화명"] == k][["영화명", "감독"]])
```

```python
url = "https://ko.wikipedia.org/wiki/" + "겨울왕국2"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

# 제작사 정보가 있는 Infobox 테이블의 label 태그를 선택
infobox_label = soup.select('.infobox-label')

# 제작사라는 텍스트를 가진 label 태그의 자손 th 태그 선택
''.join([label.find_next('a').text for label in infobox_label if label.text == '제작사'])
```

```python
production_company_list = {}

for title in tqdm(제작사):
    try:
        url = "https://ko.wikipedia.org/wiki/" + title
        res = requests.get(url)
        soup = BeautifulSoup(res.text, "html.parser")
        # 제작사 정보가 있는 Infobox 테이블의 label 태그를 선택
        infobox_label = soup.select('.infobox-label')

        # 제작사라는 텍스트를 가진 label 태그의 자손 th 태그 선택
        production_company_list[title] = ''.join([label.find_next('a').text for label in infobox_label if label.text == '제작사'])
    except:
        pass
```

```python
for k, v in production_company_list.items():
    if v != '':
        m.loc[m["영화명"] == k, "제작사"] = v
for index, row in m.iterrows():
    print(row['영화명'], row['제작사'])
```

```python
production_company_list_1 = {}

for title in tqdm(제작사):
    try:
        url = "https://ko.wikipedia.org/wiki/" + title + " (영화)"
        res = requests.get(url)
        soup = BeautifulSoup(res.text, "html.parser")
        # 제작사 정보가 있는 Infobox 테이블의 label 태그를 선택
        infobox_label = soup.select('.infobox-label')

        # 제작사라는 텍스트를 가진 label 태그의 자손 th 태그 선택
        production_company_list_1[title] = ''.join([label.find_next('a').text for label in infobox_label if label.text == '제작사'])
    except:
        pass
```

```python
for k, v in production_company_list_1.items():
    if v != '':
        m.loc[m["영화명"] == k, "제작사"] = v
for index, row in m.iterrows():
    print(row['영화명'], row['제작사'])
```
